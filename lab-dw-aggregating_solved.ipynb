```python
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
#Datacleaning
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.drop(columns=['unnamed:_0'], inplace=True)
df['gender'] = df['gender'].replace({'female': 'F', 'Femal': 'F', 'F': 'F','Male': 'M', 'M': 'M'})
df['state'] = df['state'].replace({'AZ': 'Arizona', 'Cali': 'California', 'WA': 'Washington', 'NV': 'Nevada','OR': 'Oregon'})
df['education'] = df['education'].replace('Bachelors', 'Bachelor')
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(str).str.replace('%', '').astype(float)
df['vehicle_class'] = df['vehicle_class'].replace({'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury'})
df['number_of_open_complaints'] = df['number_of_open_complaints'].astype(str).str.split('/').str[1]
#rounding all numercial type daat
df[df.select_dtypes(include=['float64', 'int64']).columns] = df.select_dtypes(include=['float64', 'int64',]).round(2)
#dropping Nan rows and repeats
df.dropna(how="all",inplace=True)
df.drop_duplicates
```

#filling Nan values

```python
#filling categorical variables with mode 
df['state'] = df['state'].fillna(df['state'].mode()[0])
df['vehicle_class'] = df['vehicle_class'].fillna(df['vehicle_class'].mode()[0])
df['vehicle_size'] = df['vehicle_size'].fillna(df['vehicle_size'].mode()[0])
df['response'] = df['response'].fillna(df['response'].mode()[0])
#vehicle type has one value
```

```python
#filling numercial values
df['months_since_last_claim'] = df['months_since_last_claim'].fillna(df['months_since_last_claim'].mean()).round(2)
df['number_of_open_complaints'] = df['number_of_open_complaints'].fillna(df['number_of_open_complaints'].mean()).round(2)
```

```python
df.isnull().sum()
```

```python
df['number_of_open_complaints'].value_counts()
```

```python
#all number_of_open_complaints values are Nan, dropping column
df.drop(columns=['number_of_open_complaints'], inplace=True)

#filling vehicle type with unknown instead of dropping them to save more data for analysing
df['vehicle_type'] = df['vehicle_type'].fillna('unknown')
```

#3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
total_number_customers_group= df.groupby(['policy_type', 'state'])['customer'].count()
customer_group=total_number_customers_group[total_number_customers_group>500]
customer_group
```

```python
#or another solution
total_customers= df.groupby(['policy_type', 'state']).filter(lambda x: x['customer'].count() > 500)
total_customers.head()
```
